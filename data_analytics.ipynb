Documento Python notebook com analise de dados referentes a emissão de plasticos nos oceanos em 2019 e 2015.

Importando bibliotecas:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Leitura de dataframe com dados sobre a quantidade de plasticos que países emitiram no oceano no ano de 2019
data_global_plastic_2019 = dataset_plastics_2019 = pd.read_csv(r"data\global-plastic-waste-emitted-to-the-ocean.csv")
```

```python
data_global_plastic_2019
```

```python

def ExtrairContinentes(data):
    """
    Função para extrair os dados apenas do continenes, usando os dataframes como parametro
    """
    df_continentes = data.loc[
    (data['Entity'] == 'Africa') | (data['Entity'] == 'Asia') |
    (data['Entity'] == 'Europe') | (data['Entity'] == 'North America') |
    (data['Entity'] == 'South America') | (data['Entity'] == 'Oceania') |
    (data['Entity'] == 'Central & North America') 
]
    return(df_continentes)
```

```python
# Aplicando a função ao data frame df_continentes_2019
df_continentes_2019 = ExtrairContinentes(data_global_plastic_2019)
df_continentes_2019
```

```python
# Plotando gráfico de barra sobre a relação de quantidade em porcentagem(x) de plástico emitida pelos continentes(x)
x = df_continentes_2019["Entity"]
y = df_continentes_2019['Share of global plastics emitted to ocean']
plt.bar(x,y)
plt.title("Quantidade de plástico emitido pelos contientes em 2019 (Em porcentagem)")
```

```python
# Criando dataframe apenas com países, excluíndo os continentes
df_paises = data_global_plastic_2019.loc[~(data_global_plastic_2019.index.isin(df_continentes_2019.index))]
df_paises
```

```python
#Criando dataframe com dados sobre a emição de plastico nos oceanos pelos continetes e diversos países
data_global_plastic_2015 = dataset_plastics_2019 = pd.read_csv(r"data\plastic-top-20-rivers.csv")
```

```python
data_global_plastic_2015
```

```python
#Criando dataframe epenas com os continentes do data_global_plastic_2015
df_continentes_2015 = ExtrairContinentes(data_global_plastic_2015)
df_continentes_2015 
```

```python
#Plotando gráfico entre a relação dos continentes(x) e a quantidade de plático que emitem em toneladas(y)
x = df_continentes_2015["Entity"]
y = df_continentes_2015['Plastic mass input from rivers (tonnes)']
plt.bar(x,y)
plt.title("Quantidade de plástico despejada nos rios em 2015 (Em toneladas)")
```

```python
#Somando a quantidade de toneladas registradas em 2015 no dataframe
lista_valores = []
for index, row in data_global_plastic_2015.iterrows():
    valor = row['Plastic mass input from rivers (tonnes)']
    lista_valores.append(int(valor)) 

soma = sum(lista_valores)
print(f"Quantidade total: {soma}")
```
